```python
# Define the names of the dependent variables
dependent_variables = ['charges']  # Replace with the actual dependent variable names

# Iterate through the list of dependent variables
for dependent_variable in dependent_variables:
    # Identify the data type of the current dependent variable
    dependent_variable_type = data[dependent_variable].dtype

    # Identify the independent variables and their data types
    independent_variables = [col for col in data.columns if col not in dependent_variables]
    independent_variable_types = data[independent_variables].dtypes

    # Display the results for each dependent variable
    print(f"\nDependent Variable: {dependent_variable} - Data Type: {dependent_variable_type}")
    print("\nIndependent Variables:")
    print(independent_variable_types)
```

```python
import pandas as pd

# Assuming 'data' is your DataFrame

# Compute summary statistics for continuous variables
continuous_variables = [col for col in data.columns if data[col].dtype != 'object']
summary_continuous = data[continuous_variables].describe()

# Display summary statistics for continuous variables
print("Summary Statistics for Continuous Variables:")
print(summary_continuous)

# Display frequency distribution for categorical variables
categorical_variables = [col for col in data.columns if data[col].dtype == 'object']
for col in categorical_variables:
    print(f"\nFrequency distribution for {col}:")
    print(data[col].value_counts())
```
